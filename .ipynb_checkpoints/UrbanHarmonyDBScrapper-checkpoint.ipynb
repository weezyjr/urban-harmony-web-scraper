### Decelartions

```python
http = urllib3.PoolManager()
```

## Scrap URL

```python
def scrapUrl(url):
    #request the url
    response = http.request('GET', url)
    
    #decode to windows arabic
    response = response.data.decode('Windows-1256')
    
    #soup the response
    return BeautifulSoup(response, 'html.parser')
```

## Scrap Image

```python
def getImgFromUrl(url):
    soup = scrapUrl(url)
    img = soup.find('img')
    return SITE_URL + img.get('src')
```

## Scrap Urban Harmony Data

```python
def scrapUrbanHarmony(url,start,finish):
    for i in range(start,finish):
        #scrap urls between 1 to 206
        soup = scrapUrl(url + str(i))

        #initilize rows array
        rows = []

        #loop through tr in table ignoring the title row
        for row in soup.find_all('tr')[10:]:
            #get all the data in row
            data = row.find_all('td')
            #skip empty data
            if len(data) == 0: 
                break
            else:
                #save the attribute in variables
                name = ''
                id = data[0].get_text(strip=True)
                type = data[1].get_text(strip=True)
                addrs = data[2].get_text(strip=True)
                value = data[3].get_text(strip=True)
                owner = data[4].get_text(strip=True)
                
                #in case there is name attribute
                try:
                    if(data[6].find('a') != None):
                        name = data[5].get_text(strip=True)
                        lawsuit_url = data[6].find('a')
                    else:
                        lawsuit_url = data[5].find('a')
                except:
                    lawsuit_url = data[5].find('a')

                #in case there is a link
                if lawsuit_url != None:
                    lawsuit_url = SITE_URL + lawsuit_url.get('href')
                    img_url = getImgFromUrl(lawsuit_url)
                    img_url = img_url.strip().replace(' ', '%20')
                else:
                    lawsuit_url = ''
                    img_url = ''
                    
            #append the data to the rows
            rows.append([id,type,addrs,value,name,owner,img_url,i])

        #write to the csv file each row if exist
        writer.writerows(row for row in rows if row)

        #print current page id
        print('current page', i)
```

## Scrap Urban Harmony Link 1

```python
scrapUrbanHarmony(LINK1,1,206)
```

## Scrap Urban Harmony Link 2

```python
scrapUrbanHarmony(LINK2,1,206)
```
