```python
import urllib3
```

```python
import csv
```

```python
from bs4 import BeautifulSoup
```

```python
SITE_URL = 'http://urbanharmony.org/'
```

```python
LINK = SITE_URL + "valueplaces.asp?id="
LINK2 = SITE_URL + "valueplaces2.asp?id="
LINK3 = SITE_URL + "valueplaces3.asp?id="
PROVLINK = SITE_URL + "ar_cities.asp"
```

```python
http = urllib3.PoolManager()
```

```python
file = open('places.csv', 'w', newline='')
```

```python
writer = csv.writer(file)
```

```python
writer.writerow(['page','prov_id','prov','city']);
```

```python
def getLinks(link):
     #request links between 1 to 206
    response = http.request('GET', link)

    #decode to windows arabic
    response = response.data.decode('Windows-1256')

    #soup the response
    soup = BeautifulSoup(response, 'html.parser')

    #return links
    return soup.find_all('a')
```

```python
def getProvs():
    provs = []
    #get links
    data = getLinks(PROVLINK)[25:]
    for a in data:
        if a.get('href') != None:
            provs.append([a.get_text(strip=True),a.get('href').split('=')[1]])
    return provs
```

```python
def getAreas(link):
    areas = []
    for i in range(1,29):
        #get links
        data = getLinks(link + str(i))[14:]
        for a in data:
            if a.get('href').startswith('placedetails.asp?id=') == True:
                areas.append([i,a.get_text(),a.get('href').split('=')[1]])
    return areas
```

```python
provs = getProvs()
```

```python
areas1 = getAreas(LINK)
```

```python
areas2 = getAreas(LINK2)
```

```python
areas = getAreas(LINK3)
```

```python
areas = areas1 + areas2
```

```python
for area in areas:
    for prov in provs:
        if(int(area[0]) == int(prov[1])):
            area.append(prov[0])
```

```python
for area in areas:
    area = [area[2],area[0],area[1],area[3]]
    writer.writerow(area)
    print(area)
```

```python
file.close()
```
